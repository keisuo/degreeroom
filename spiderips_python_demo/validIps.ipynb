```python
import re
import requests
import threading

srcfile = open('proxies.txt', 'r')
outfile = open('verified.txt', 'w')

url = "http://esf.gz.fang.com/map/?mapmode=y&district=78&subway\
line=&subwaystation=&price=0-170&room=&area=&towards=&floor=&hage=&equipment=&keyword=%E5%B8%82%E6%A1%A5%E4%B8%AD%E5%BF%83%E5%B0%8F%E5%AD%A6&comarea=&orderby=16&isyouhui=&newCode=&ho\
useNum=&schoolDist=&schoolid=&ecshop=&groupedmode=4&PageNo=1&zoom=16&a=ajaxSearch&city=gz&searchtype=&housetag=81"
user_agent = "Mozilla/5.0 (Windows NT 10.0; WOW64; rv:52.0) Gecko/20100101 Firefox/52.0"
headers = {'user-agent': user_agent}

mutex = threading.Lock()

def verify():
    try:
        mutex.acquire()
        proxy = srcfile.readline().strip()
        mutex.release()
        
        proxies = {'http': proxy}
        
        r = requests.get(url, proxies = proxies, headers = headers, timeout = 5)
        if(r.status_code==200):
            print(proxy)
            mutex.acquire()
            outfile.write('\'%s\'\n' % proxy.replace('http://',''))
            mutex.release()

    except requests.RequestException as e:
        pass

srcfilet = open('proxies.txt', 'r')
lines = len(srcfilet.readlines()) 
srcfilet.close()

childthread = []

for i in range(lines):
    t = threading.Thread(target = verify)
    childthread.append(t)
    t.start()

for t in childthread:
    t.join()


srcfile.close()
outfile.close()
print('finish')
```
